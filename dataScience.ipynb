The project is an analysis of an exoplanet dataset from Kaggle Datasets. The dataset is basically a list of discovered planets in a span of the past few centuries. It contains of roughly 3584 rows plus a header meaning that it has roughly 3584 exoplanets. The aim is to analyze the dataset features, which include, Planetary mass, Planet radius, Planet surface Temperature, Host star Temperature, Semi major axis, Orbital periods, Discovery method, Discovery year, Host Star mass, Eccentricity, Host star metallicity among other features. With cleaning, analysis and visualization; one can be able to describe correlations between specific features like Host Star Temperature and Planet surface Temperature to answer various questions like "Is the Temperature of a planet determined by the temperature of it`s Host star?"

The code in the following code cell, imports all the libraries I will need in the analysis process of the data set. Pandas library is for handling the data (reading, describing and  creating data frames where needed and more), Matplotlib is for effective visualization upon cleaning and analysis of the data and seaborn will be used to provide more features in areas where Matplotlib features are not enough.


```python
import  pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

The following code will access the dataset. After reading the data, proceed to find the columns that have the most missing values and also convert this into percentage so as to  get the perspective of how much data is missing. After that,  create a new table from a dictionary and combine the missing values count and the missing percentage which is rounded to one decimal place for neatness.
Then  proceed to print the result in an order such that the columns with the most missing values are on top.

```python

data = pd.read_csv("oec.csv")

missing = data.isnull().sum()
missing_percentage = (data.isnull().sum() / len(data)) * 100
missing_summary = pd.DataFrame({
    "missing count": missing,
    "missing percentage": missing_percentage.round(1)
}
)
print (missing_summary.sort_values('missing percentage', ascending=False))
```

After printing the columns with most missing values,  proceed to analyze them to determine their usability. For the columns that are missing too much data ranging from 80% and above, they are  dropped since they cannot be used.
The code in the cell below is for this purpose, create a list and in here, store the columns that turned out to miss a lot of data.  After that,  create a new variable (cleaned_data) which stores the new cleaned data. Keep with the original data in case its needed somewhere else. After this output the cleaned data to analyze it.

```python

dropped_cols = [
    "AgeGyr", "LongitudeDeg", "AscendingNodeDeg",
    "PeriastronDeg", "HostStarAgeGyr", "InclinationDeg",
    "ListsPlanetIsOn","LastUpdated", "TypeFlag",
    "RightAscension", "Declination",
]

cleaned_data = data.drop(columns=dropped_cols)
print(cleaned_data.shape)
print (cleaned_data.columns.tolist())
print (cleaned_data.describe())
```

In the following code cell, the code visualizes  the methods by which planets were discovered. First  get total count of the discovery methods in the data. After that use matplotlib to visualize using bar charts. The output here indicates that most of the planets were discovered through Transit method( detect planets by their orbiting movements in front of their star which creates dips in the host star's brightness), then RV(radio velocity) which is detection of tiny wobbles a planet causes in it' s star as it orbits, then imaging which is directly photographing planets and then microlensing and timing.

```python

discovery_counts = cleaned_data ["DiscoveryMethod"].value_counts()
plt.figure(figsize =(10,6))
plt.bar (discovery_counts.index, discovery_counts.values, color = "steelblue", edgecolor = "white")
plt.title('Exoplanet Discovery methods')
plt.xlabel('Discovery method')
plt.ylabel('Number of Planets ')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

In the cell code below , the code visualizes planet discoveries per year.  The index gets the particular year amd the value gets the number of planets discovered that particular year. Upon visualizing  notice that most planets were discovered starting from 2009 going upwards. The highest number of planets discovered ranged from (2014 - 2016).

```python

discoveries_per_year = cleaned_data["DiscoveryYear"].value_counts().sort_index()
plt.figure(figsize =(12,6))
plt.plot(discoveries_per_year.index, discoveries_per_year.values,
         color = "cyan", linewidth = 2, marker = "o", markerfacecolor = "red", markeredgecolor = "black", markersize = 4)
plt.title("Exoplanets Discovery per Year")
plt.xlabel('Discovery year')
plt.ylabel('Number of Planets ')
plt.grid(True, alpha = 0.4, which = "major")
plt.minorticks_on()
plt.grid(True, alpha = 0.2, which = "minor")
plt.tight_layout()
plt.show()
```

The code cell below shows the visualization of Planetary mass versus the Radius of the planet. The code, checks the cleaned data set and searches for missing values in Planetary mass and Radius columns and drops these rows, since we cannot visualize rows that are missing either of the data or both. After dropping the rows with missing values, plot the graph using scatter plots.
Notice that due to the presence of extreme outliers the scatter plot will be compacted and visual analysis will not be as effective, therefore  use (plt.xlim and plt.ylim) to specify which ranges of the visualized data  to look at. After visualization,  notice that most planets ranged in (0 - 3 jupiter masses and 0- 1.9 jupiter radii), while not entirely zero most planets' radii was very small values. Also notice that some planets are very heavy (about 30 Jupiter masses ) but their radii are small, while others have huge radii but their masses are less. This all comes down to the composition elements making up the planet. Some are small (less radius) but they are made up of dense rocky materials hence the mass, while others are gaseous giants bloated due to their stars heat, but gases are lighter, hence the smaller mass.

```python

mass_radius = cleaned_data.dropna(subset=["PlanetaryMassJpt", "RadiusJpt"])
plt.figure(figsize =(10,6))
plt.scatter(mass_radius["PlanetaryMassJpt"], mass_radius["RadiusJpt"], color ="red", alpha = 1, edgecolor = "white", s = 5, linewidth = 0.5)
plt.title("Planetary Mass vs Radius")
plt.xlabel("Planetary Mass(Jupiter Masses)")
plt.ylabel("Radius (Jupiter Radii)")
plt.minorticks_on()
plt.grid(True, alpha = 0.4, which = "major")
plt.grid(True, alpha = 0.2, which = "minor")
plt.xlim(0, 6) and plt.ylim(0, 3)
plt.tight_layout()
plt.show()
print (mass_radius["PlanetaryMassJpt"].describe())
print (mass_radius["RadiusJpt"].describe())
```

In the code cell below, the code visualizes the distribution of Orbital Periods of the Exoplanets. First step is dropping the rows with missing Period data. After the rows are successfully dropped,  proceed to create a histogram, showing the where the Orbital periods of the planets range. Notice that, most planets (about 2100 planets) have Orbital periods ranging between (0 to 50 days), which means some have their "year"  smaller than a month on Earth. This indicates that the planets must orbit very close to their parent star. This bias results from the discovery methods of  planets as most are discovered through Transit method (detection of planets, through their orbiting behavior in front of their stars that means the star's brightness occasionally dips when blocked by the orbiting planet). Additionally, "hot Jupiters" are common (gaseous giants that orbit very close to their star) meaning that they complete their orbital periods in days.

```python

period_data = cleaned_data.dropna(subset=["PeriodDays"])
period_data = period_data[period_data["PeriodDays"] < 1000]
plt.figure(figsize =(10,6))
plt.hist(period_data["PeriodDays"], bins = 50, color = "purple", edgecolor = "white")
plt.title ("Distribution of Orbital Periods")
plt.xlabel("Orbital Period(Days)")
plt.ylabel("Number of Planets ")
plt.minorticks_on()
plt.grid(True, alpha = 0.4, which = "major")
plt.grid(True, alpha = 0.2, which = "minor")
plt.tight_layout()
plt.show()
```

In the following cell, the code  visualizes Host star temperature against planet surface temperature. We want to find out whether hotter stars produce hotter planets. As usual, the first step involves dropping rows with missing values. After that,  plot the graph using scatter plots. After visualization, notice that most stars temperature range from (2500 - 7000K) while planets range from (200 - 2600K). However, one planet had an odd temperature of about 7100K, which is very hot for a planet. The planet turns out to be KOI - 55b, which from research is said to have survived being "eaten" by it' s parent star,  now it orbits around the surface of the star thus the high temperature. The hottest star in the data set has a temperature of 30,000K , but the planet orbiting it is surprisingly cool, which means it must orbit the star at a very far distance. Therefore, we can conclude that hotter stars do not particularly produce hotter planets, rather, the distance is a key factor for the planet's temperature among other factors such as thick atmosphere and the heat that some planets generate from inside.

```python

temp_data = cleaned_data.dropna(subset=["HostStarTempK", "SurfaceTempK"])
plt.figure(figsize =(10,6))
plt.scatter(temp_data["HostStarTempK"], temp_data["SurfaceTempK"], color ="blue", alpha = 1, edgecolor = "white")
plt.title("Host Star Temperature vs Planet Surface Temperature")
plt.xlabel("Host Star Temperature (K)")
plt.ylabel("Planet Surface Temperature (K)")
plt.minorticks_on()
plt.grid(True, alpha = 0.6, which = "major")
plt.grid(True, alpha = 0.4, which = "minor")
plt.tight_layout()
plt.show()
```

The final code cell, shows the correlation of all the features of the cleaned data. Seaborn is useful in order to create a heatmap visualization of these features, showing strong positive, no correlation or zero and strong negative.

```python

numerical_data = cleaned_data.select_dtypes(include = "number")
correlation = numerical_data.corr()
plt.figure(figsize =(10,6))
sns.heatmap(correlation,
            annot=True,
            fmt=".2f",
            cmap="coolwarm",
            center=0,
            linewidths=0.5,)
plt.title("Correlation Heatmap Of Exoplanet Features")
plt.tight_layout()
plt.show()
```

The code below plots Semi Major Axis (Astronomical Units) against Planet Surface Temperature. This results show that most hot planets are orbiting their planet star at a close distance.
However, there is one outlier, a planet orbiting it ' s star  at a distance of  662AU and has a temperature of 1600K. This seems weird because at that distance, the planet should be near absolute zero. The planet turns out to be, (CVSO 30 c). CVSO is a very young star system, CVSO 30 c is very hot because its still glowing from its own formation. Thus, this explains the extreme temperatures at an extremely far distance like that.

```python
temp_axis = cleaned_data.dropna(subset=["SurfaceTempK", "SemiMajorAxisAU"])
plt.figure(figsize =(10,6))
plt.scatter(temp_axis["SurfaceTempK"], temp_axis["SemiMajorAxisAU"], color ="purple", alpha = 1, edgecolor = "white")
plt.title("Surface Temperature vs Semi Major Axis ")
plt.xlabel("Surface Temperature (K)")
plt.ylabel("Semi Major Axis (AU)")
plt.minorticks_on()
plt.grid(True, alpha = 0.4, which = "major")
plt.grid(True, alpha = 0.3, which = "minor")
plt.tight_layout()
plt.show()
```
